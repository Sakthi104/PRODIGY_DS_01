```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("population_csv.csv")
```

```python
df.head(7)
```

```python
df.dropna(inplace=True)
```

```python
df.columns=df.iloc[0]
```

```python
df=df[1:]
```

```python
df.reset_index(inplace=True)
df.drop(['index','Country Name'	,'Country Code'	,'Indicator Name',	'Indicator Code'	],axis=1,inplace=True)
```

```python
sum=pd.DataFrame(df.sum(axis=0))
sum=pd.DataFrame(sum.floordiv(1000000000))
```
